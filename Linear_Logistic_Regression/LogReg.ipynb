```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd

df_train = pd.read_csv("Titanic_train.csv")
df_test = pd.read_csv("Titanic_test.csv")
df_Y_test = pd.read_csv("Titanic_submission_check.csv")

print(df_test.shape)
print(df_train.shape)
print(df_train.head())
print("---------------------------------------")
print(df_test.head())
```

```python
print(df_train.info())
print(df_test.info())
```

```python
cols_rem = ["Name", "Ticket", "Cabin", "Embarked"]
df_train = df_train.drop(columns=cols_rem)
df_test = df_test.drop(columns=cols_rem)
print(df_train.head())
print(df_test.head())
print(df_test.shape)
print(df_train.shape)
```

```python
print(df_train.isna().sum())
print(df_test.isna().sum())
```

```python
df_train = df_train.dropna()
df_test["Age"] = df_test["Age"].fillna(df_test["Age"].median())
df_test["Fare"] = df_test["Fare"].fillna(df_test["Fare"].median())
print(df_train.isna().sum())
print(df_test.isna().sum())

print(df_train.shape)
print(df_test.shape)
```

```python
df_train["Sex"] = df_train["Sex"].replace({"male":0, "female":1})
df_test["Sex"] = df_test["Sex"].replace({"male":0, "female":1})
print(df_train.head())
print(df_test.head())
```

```python
X_cols = list(df_train.columns)
X_cols.remove("Survived")
X_cols.remove("PassengerId")
X_train = df_train[X_cols]
X_test = df_test[X_cols]
Y_train = df_train["Survived"]
Y_test = df_Y_test["Survived"]
```

```python
reg = LogisticRegression()
reg.fit(X_train, Y_train)
```

```python
Y_pred = reg.predict(X_test)
print(Y_pred)
```

For the Titanic Logistic Regression task:

To figure out if a passenger survived I used a set of modules from the scikit-learn library. 
- The main model I used is **sklearn.linear_model.LogisticRegression**. Even though it has regression in its name this is actually a classification algorithm that calculates the probability of a data point belonging to a category, such as survived or did not survive, by fitting the data to a logistic S-shaped Sigmoid curve.  
- Finally to see how well the model did I used **sklearn.metrics.accuracy_score**. This metric is for classification tasks and calculates the fraction of correctly classified samples out of the total number of predictions the model made on the test set.

- Regarding the differences in the evaluation metrics:
It is important to know the difference between the metrics we use for these two types of tasks. For regression, which predicts continuous numbers like prices we use Mean Squared Error to see how far off our predictions are and we square the errors so big mistakes are more pronounced. We also use the R-squared score to see how well the model explains the variation in the data. For this score a number closer to 1.0 is better. These metrics do not work for classification tasks like the Titanic Logistic Regression task. Instead for regression, which predicts discrete categories, like survived or died we use the accuracy score. This is a metric that shows the total number of correct guesses divided by the total number of guesses so a higher percentage means the model is doing better. The Titanic Logistic Regression task uses this accuracy score to evaluate the models performance.
