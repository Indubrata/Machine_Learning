```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd

df = pd.read_csv("HousingData.csv")
print(df.head())
print(df.describe())
print(df.info())
print(df.shape)
```

```python
print(df.isnull().sum())
df = df.dropna()
print(df.isnull().sum())
```

```python
x_col = list(df.columns)
x_col.remove("MEDV")
print(x_col)
```

```python
X = df[x_col]
Y = df["MEDV"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=89)
print(f"Training set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, Y_train)
```

```python
Y_pred = reg.predict(X_test)
print(Y_pred)
```

```python
from math import sqrt
mse = sqrt(mean_squared_error(Y_test, Y_pred))
r2_sc = r2_score(Y_test, Y_pred)
print(f"The predicted median value of owner occupied homes is off by ${round(mse * 1000.0, 2)} with respect to the test data")
print(f"{round(r2_sc * 100.0, 2)}% of variation in house prices can be explained by the model")
```

For the Boston Housing Linear Regression task:

I did an analysis where I used some modules from the scikit-learn library to work with the given data teach the model and see how well it does. 
- The main machine learning algorithm I used is **sklearn.linear_model.LinearRegression**, which finds a line that best fits the training data by making the difference between what was observed and the real house prices as small as possible. 
- I wanted to make sure the model really learns and does not just remember the data so I used **sklearn.model_selection.train_test_split** to mix up the data and split it into two parts: one part to teach the model and another part to test it.
- To see how good the model is I used two ways to score it. 
- The **sklearn.metrics.mean_squared_error** function calculates the Mean Squared Error, which is the average difference between what the test data showed the house prices should be and what the model predicted it to be. 
- The **sklearn.metrics.r2_score** function calculates the R-squared value, which shows how well the model explains the house prices by telling what percentage of the differences, in house prices can be explained by the features provided in the training data.
- What I learned through the video:-
Simple Linear Regression is a way to predict a continuous value based on another value.

It helps to forecast a number using another number.

* Independent Variable ($x$): This is the input or feature used to make predictions.

For example it could be the area of a home in feet.

* Dependent Variable ($y$): This is the output or target the model is trying to predict.

For example it could be the price of the home.

The goal is to find a line that best fits the data points on a graph.

This line helps predict $y$ for any given $x$.

2. The Mathematics: How the "Best Fit" is Found

When you plot data points on a graph inf lines can be drawn through them.

The model finds the best line by reducing errors:

* The Error (Delta): This is the distance between a data point and the point predicted by the line.

* Minimizing the Error: The algorithm squares each error adds them up and adjusts the line until the sum of errors is as small as possible.

The equation of the line is: $y = mx + b$.

* $m$ (Slope/Coefficient): This shows how steep the line is.

In scikit-learn this is stored as coef_.

* $b$ (Intercept): This is where the line crosses the y-axis.

In scikit-learn this is stored as intercept_.
